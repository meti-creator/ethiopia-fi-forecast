```python
import pandas as pd

main_df = pd.read_excel("../data/raw/ethiopia_fi_unified_data.xlsx", sheet_name="ethiopia_fi_unified_data")
impact_df = pd.read_excel("../data/raw/ethiopia_fi_unified_data.xlsx", sheet_name="Impact_sheet")
ref_codes = pd.read_excel("../data/raw/reference_codes.xlsx")

print("Main sheet shape:", main_df.shape)
print("Impact sheet shape:", impact_df.shape)
print("Reference codes shape:", ref_codes.shape)

print("\nrecord_type breakdown (main sheet):")
print(main_df['record_type'].value_counts())
```

**lets counts by pillar, source_type, and confidence, plus the temporal range of observations**.

```python
print("Pillar breakdown (observations + targets only):")
print(main_df[main_df['record_type'].isin(['observation', 'target'])]['pillar'].value_counts())

print("\nCategory breakdown (events only):")
print(main_df[main_df['record_type'] == 'event']['category'].value_counts())

print("\nSource type breakdown:")
print(main_df['source_type'].value_counts())

print("\nConfidence breakdown:")
print(main_df['confidence'].value_counts())

print("\nTemporal range of observations:")
obs = main_df[main_df['record_type'] == 'observation']
print("Earliest:", obs['observation_date'].min())
print("Latest:", obs['observation_date'].max())

print("\nUnique indicators and how many times each appears:")
print(main_df['indicator_code'].value_counts())
```

```python
print("Events cataloged (sorted by date):")
events = main_df[main_df['record_type'] == 'event'][['observation_date', 'category', 'indicator', 'notes']].sort_values('observation_date')
print(events.to_string(index=False))

print("\n\nImpact links - which events have links, and to what:")
impact_summary = impact_df.merge(
    main_df[main_df['record_type']=='event'][['record_id','indicator']],
    left_on='parent_id', right_on='record_id', suffixes=('_impact','_event')
)
print(impact_summary[['indicator_event', 'pillar', 'related_indicator', 'impact_direction', 'evidence_basis']].to_string(index=False))
```
