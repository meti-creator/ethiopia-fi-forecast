```python

# ============================================================
# TASK 4 (OFFICIAL): FORECAST ACCESS AND USAGE
# Focus: Account Ownership & Digital Payment Usage
# Horizon: 2025-2027
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

print("=" * 80)
print("TASK 4: FORECASTING ACCESS AND USAGE (2025-2027)")
print("=" * 80)
print()

# Reload data
df = pd.read_excel('/mnt/agents/upload/ethiopia_fi_unified_data.xlsx')
df = df.iloc[1:].reset_index(drop=True)
df['observation_date'] = pd.to_datetime(df['observation_date'], errors='coerce')
df_impact = pd.read_excel('/mnt/agents/upload/ethiopia_fi_unified_data.xlsx', sheet_name=1)
df_impact = df_impact.iloc[1:].reset_index(drop=True)

obs = df[df['record_type'] == 'observation']
events = df[df['record_type'] == 'event']
targets = df[df['record_type'] == 'target']

# ============================================================
# STEP 1: DEFINE TARGETS
# ============================================================
print("STEP 1: Define Targets")
print("-" * 80)
print()

# Target 1: Account Ownership Rate
acc_target = targets[targets['indicator_code'] == 'ACC_OWNERSHIP']
if len(acc_target) > 0:
    print("TARGET 1: Account Ownership Rate (ACCESS)")
    print(f"  Definition: % of adults with account at financial institution or mobile money")
    print(f"  Target: {acc_target.iloc[0]['value_numeric']}% by {acc_target.iloc[0]['observation_date'].year}")
    print(f"  Source: {acc_target.iloc[0]['source_name']}")
    print()

# Target 2: Digital Payment Usage
# We need to construct this from P2P + other usage indicators
print("TARGET 2: Digital Payment Usage")
print(f"  Definition: % of adults who made or received digital payment")
print(f"  (Derived from P2P transaction count + mobile money active users)")
print()

# Get historical data for both
acc_hist = obs[(obs['indicator_code'] == 'ACC_OWNERSHIP') & (obs['gender'] == 'all')].sort_values('observation_date')
print("Historical Account Ownership:")
for _, row in acc_hist.iterrows():
    yr = row['observation_date'].year if pd.notna(row['observation_date']) else row['fiscal_year']
    print(f"  {yr}: {row['value_numeric']:.0f}%")

print()

# For digital payment usage, we'll use P2P transaction growth as proxy
# and also look at mobile money account penetration
p2p_hist = obs[(obs['indicator_code'] == 'USG_P2P_COUNT') & (obs['gender'] == 'all')].sort_values('observation_date')
mm_hist = obs[(obs['indicator_code'] == 'ACC_MM_ACCOUNT') & (obs['gender'] == 'all')].sort_values('observation_date')

print("Historical Digital Payment Proxies:")
print("  P2P Transactions:")
for _, row in p2p_hist.iterrows():
    yr = row['observation_date'].year if pd.notna(row['observation_date']) else row['fiscal_year']
    print(f"    {yr}: {row['value_numeric']/1e6:.1f}M")
print("  Mobile Money Accounts:")
for _, row in mm_hist.iterrows():
    yr = row['observation_date'].year if pd.notna(row['observation_date']) else row['fiscal_year']
    print(f"    {yr}: {row['value_numeric']:.1f}%")

print()
print("=" * 80)
print("✅ STEP 1 COMPLETE: Targets defined")
print("=" * 80)
```
